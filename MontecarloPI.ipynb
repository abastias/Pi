**The Montecarlo approach**. In order to apply this method using the Montecarlo simulation, we need to have a set of point (x,y) and we need to calculate the aproximation for the probability. This probabilty P is obtained dividing the *cases inside the circle by the total of iteration*. Remember, our square is -1 to 1, and the circle has a radio r=1. In order to know if a par(x,y) is inside the cicle  We need to calculate the distance to the center of the square, the coordenates (0,0). If the distance is less or equal to 1, the point in inside the circle. Using Pythagoras theorem, we calculate the distance *D* as

Steps:
1. Generate a set(x,y) of uniform random number from -1 to 1. This set will identify a position inside the square.
2. Identify the number of cases that are inside the circle.
3. Calculate the probabilty of the any set (x,y) will be inside the circle and estimate Pi

```python
# Basic libraries to run the process
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the numbers of Iterations  N, the bigger the better
N = 100000
```

```python
#Step 1. Generate a set(x,y) of uniform random number from -1 to 1. This set will identify a position inside the square.
x = np.random.uniform(low=-1, high=1, size=[N,1])
y = np.random.uniform(low=-1, high=1, size=[N,1])
```

```python
#2. Identify the number of cases that are inside the circle.
insideCircle = x**2+y**2 <=1  # Boolean value
```

```python
#3. Calculate the probabilty of the any set (x,y) will be inside the circle and estimate Pi

calc_Pi= 4*np.sum(insideCircle)/N

```

```python
# To makes this process more visual, we can plot all point, green will be inside the circle, red outside

x_insideCircle= x[insideCircle]
y_insideCircle = y[insideCircle]

plt.figure (figsize=[5,5])
plt.scatter (x,y, color='r',s=1)
plt.scatter(x_insideCircle,y_insideCircle, color='g',s=1)
plt.show()

```

```python

print('Pi: {}, Calc_Pi: {}, Error: {} %'.format(np.pi, calc_Pi,(np.abs(np.pi-calc_Pi)/np.pi*100 )))
# Because we are using a Montecarlo Simulation the Calculated Pi and Error can be different every time 
# we run the simulation , ven if we use the same numeber of iterations N.
```
